# Ploynomial Regression

```python
# divide dataset as independent and dependent variables

x = ds[["Area"]]
y= ds[["Price"]]
```

```python
# split dataset into Training and Testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size =0.2 ,random_state=1)
x_train
```

```python
# fiiting  Ploynimial Regression model

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=5)
x_poly =poly_reg.fit_transform(x)
x_poly
reg =LinearRegression()
reg.fit(x_poly,y)
```

```python
# predict
y_pre= reg.predict(poly_reg.transform(x))
y_pre
```

```python
# checking accuracy

from sklearn.metrics import r2_score
score = r2_score(y,y_pre)
score
```

```python
# visulization

plt.scatter(x,y)
plt.plot(x,reg.predict(poly_reg.transform(x)),color='pink')
plt.show()
```

```python
# for user input
area =float(input("Enter Area : "))
myinput =pd.DataFrame(columns =['Area'],data=[[area]])
result =reg.predict(poly_reg.transform(myinput))
print(f"Predicted Price House of Area {area} : {round(result[0,0])}")
```

# Decision Tree Regression


```python
# fitting Decision Tree Regression Model

from sklearn.tree import DecisionTreeRegressor,plot_tree
reg= DecisionTreeRegressor(random_state =1)
reg.fit(x,y)
```

```python
# predict
y_pre =reg.predict(x)
y_pre
```

```python
# checking accuracy

from sklearn.metrics import r2_score
score = r2_score(y,y_pre)
score
```

```python
# visualize decision tree

plt.figure(figsize=(50,50))
plot_tree(reg,feature_names=["Area"], filled=True)
plt.title("Decision Tree Viualization ")
plt.show()
```

```python
# for user input

area = int(input("Enter Area : "))
myinput = pd.DataFrame(columns = [["Area"]],data =[[area]])
result = reg.predict(myinput)
print(f"Price House of Area {area} : {result[0]}")
```

# Random Forest Regression

```python
# fitting Random Forest Regression

from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor(n_estimators=10,random_state =1)
reg.fit(x,y)
```
